```python
import pandas as pd
```

```python
#Carregar o arquivo excel e colocar dentro de uma váriavel 
df = pd.read_excel('dados_produtos.xlsx')
```

```python
df
```

```python
# Fazer o Tratamento dos dados onde só quero pegar as Placas de Vídeo, visto que tem produtos misturados
produtos_placa_video = df[df['Nome do Produto'].str.startswith('Placa de Vídeo')]
```

```python
# Verificar se todos os produtos são placa de vídeo
todos_sao_placa_video = produtos_placa_video['Nome do Produto'].str.startswith('Placa de Vídeo').all()
```

```python
# Exibir o resultado
if todos_sao_placa_video:
    print("Todos os produtos são de Placa de Vídeo.")
else:
    print("Existem produtos que não são de Placa de Vídeo na variável produtos_placa_video.")
```

```python
tratado = 'arquivo_tratado.xlsx'
caminho_arquivo = tratado  # O arquivo será salvo na pasta atual
produtos_placa_video.to_excel(caminho_arquivo, index=False)
```
